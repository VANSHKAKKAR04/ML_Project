```python


# 1. UNZIP THE DATASET
# --------------------
import zipfile
import os
import pathlib

# Adjust this path if your ZIP is elsewhere
ZIP_PATH    = '/content/archive (6).zip'
EXTRACT_DIR = '/content/satellite_dataset'

os.makedirs(EXTRACT_DIR, exist_ok=True)
with zipfile.ZipFile(ZIP_PATH, 'r') as zip_ref:
    zip_ref.extractall(EXTRACT_DIR)

# After unzipping, dataset structure is:
# /content/satellite_dataset/data/<class_name>/*.jpg/png
BASE_DIR = pathlib.Path(EXTRACT_DIR) / 'data'
print("Classes and image counts (before split):")
for folder in sorted(os.listdir(BASE_DIR)):
    folder_path = BASE_DIR / folder
    if folder_path.is_dir():
        count = len(list(folder_path.iterdir()))
        print(f"  • {folder}: {count} images")

# 2. SPLIT INTO TRAIN/VALIDATION/TEST (80%/10%/10%)
# -------------------------------------------------
import random
import shutil

# Set random seed for reproducibility
random.seed(42)

# Paths for new splits
TRAIN_DIR = pathlib.Path(EXTRACT_DIR) / 'train'
VALID_DIR = pathlib.Path(EXTRACT_DIR) / 'validation'
TEST_DIR  = pathlib.Path(EXTRACT_DIR) / 'test'

# Create folders for each split and each class
for split_dir in [TRAIN_DIR, VALID_DIR, TEST_DIR]:
    for cls in os.listdir(BASE_DIR):
        cls_src = BASE_DIR / cls
        if cls_src.is_dir():
            (split_dir / cls).mkdir(parents=True, exist_ok=True)

# For each class, shuffle images and move 10% to test, 10% of remainder to validation, rest to train
for cls in os.listdir(BASE_DIR):
    cls_path = BASE_DIR / cls
    all_images = list(cls_path.iterdir())
    random.shuffle(all_images)
    n_total = len(all_images)
    n_test  = int(0.10 * n_total)
    n_valid = int(0.10 * n_total)  # 10% for validation
    # Note: 80% remains for training

    test_imgs  = all_images[:n_test]
    valid_imgs = all_images[n_test : n_test + n_valid]
    train_imgs = all_images[n_test + n_valid :]

    # Move files
    for img in test_imgs:
        shutil.move(str(img), str(TEST_DIR  / cls / img.name))
    for img in valid_imgs:
        shutil.move(str(img), str(VALID_DIR / cls / img.name))
    for img in train_imgs:
        shutil.move(str(img), str(TRAIN_DIR / cls / img.name))

# After moving, original BASE_DIR/<class> should be empty; ignore that.
print("\nAfter split, counts:")
for split_name, split_dir in [('train', TRAIN_DIR), ('validation', VALID_DIR), ('test', TEST_DIR)]:
    print(f"\n{split_name.upper()} set:")
    for cls in sorted(os.listdir(split_dir)):
        cls_split_path = split_dir / cls
        count = len(list(cls_split_path.iterdir()))
        print(f"  • {cls}: {count} images")

# 3. CREATE IMAGE DATA GENERATORS
# --------------------------------
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

CLASSES       = sorted([d.name for d in TRAIN_DIR.iterdir() if d.is_dir()])
NUM_CLASSES   = len(CLASSES)
IMAGE_SIZE    = (224, 224)
BATCH_SIZE    = 32

# Training generator with augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)

# Validation & Test generators (no augmentation, only rescale)
valid_datagen = ImageDataGenerator(rescale=1.0/255.0)
test_datagen  = ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow_from_directory(
    directory=str(TRAIN_DIR),
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    classes=CLASSES,
    class_mode='categorical',
    shuffle=True,
    seed=42
)

validation_generator = valid_datagen.flow_from_directory(
    directory=str(VALID_DIR),
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    classes=CLASSES,
    class_mode='categorical',
    shuffle=False,
    seed=42
)

test_generator = test_datagen.flow_from_directory(
    directory=str(TEST_DIR),
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    classes=CLASSES,
    class_mode='categorical',
    shuffle=False,
    seed=42
)

# 4. BUILD THE NASNETMOBILE MODEL + CUSTOM HEAD
# ---------------------------------------------
from tensorflow.keras.applications import NASNetMobile
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

LEARNING_RATE = 1e-4
FINE_LR       = 1e-5
EPOCHS_HEAD   = 10
EPOCHS_FINE   = 5

# 4.1 Load pretrained NASNetMobile backbone
base_model = NASNetMobile(
    weights='imagenet',
    include_top=False,
    input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)
)
for layer in base_model.layers:
    layer.trainable = False

# 4.2 Add a pooling + dropout + softmax head
x = base_model.output
x = GlobalAveragePooling2D(name='avg_pool')(x)
x = Dropout(0.5, name='dropout')(x)
outputs = Dense(NUM_CLASSES, activation='softmax', name='predictions')(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nModel summary:")
model.summary()

# 5. TRAIN THE TOP LAYERS (HEAD-ONLY)
# -----------------------------------
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint_path = '/content/nasnet_satellite_best.h5'
checkpoint_cb = ModelCheckpoint(
    filepath=checkpoint_path,
    save_best_only=True,
    monitor='val_accuracy',
    mode='max',
    verbose=1
)
earlystop_cb = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True,
    verbose=1
)

history_head = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=EPOCHS_HEAD,
    callbacks=[checkpoint_cb, earlystop_cb]
)

# 6. FINE-TUNE THE ENTIRE MODEL
# -----------------------------
# 6.1 Unfreeze all layers
for layer in base_model.layers:
    layer.trainable = True

# 6.2 Recompile with lower learning rate
model.compile(
    optimizer=Adam(learning_rate=FINE_LR),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 6.3 Continue training
initial_epoch = len(history_head.history['loss'])
total_epochs  = EPOCHS_HEAD + EPOCHS_FINE

history_fine = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=total_epochs,
    initial_epoch=initial_epoch,
    callbacks=[checkpoint_cb, earlystop_cb]
)

# 7. PLOT TRAINING & VALIDATION CURVES
# ------------------------------------
import matplotlib.pyplot as plt

# Merge head-only and fine-tune histories
hist_loss     = history_head.history['loss'] + history_fine.history['loss']
hist_val_loss = history_head.history['val_loss'] + history_fine.history['val_loss']
hist_acc      = history_head.history['accuracy'] + history_fine.history['accuracy']
hist_val_acc  = history_head.history['val_accuracy'] + history_fine.history['val_accuracy']
epochs_range  = range(len(hist_loss))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, hist_acc, label='Train Acc')
plt.plot(epochs_range, hist_val_acc, label='Val Acc')
plt.title('Accuracy vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, hist_loss, label='Train Loss')
plt.plot(epochs_range, hist_val_loss, label='Val Loss')
plt.title('Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# 8. LOAD BEST WEIGHTS & SAVE FINAL MODEL (OPTIONAL)
# --------------------------------------------------
model.load_weights(checkpoint_path)
print(f"\n Loaded best weights from: {checkpoint_path}")

# Save as HDF5 to reuse later
final_h5_path = '/content/nasnet_satellite_final.h5'
model.save(final_h5_path, include_optimizer=False)
print(f" Final model saved to: {final_h5_path}")

# 9. EVALUATE ON TEST SET
# -----------------------
test_loss, test_acc = model.evaluate(test_generator, verbose=1)
print(f"\n Test Loss: {test_loss:.4f}")
print(f" Test Accuracy: {test_acc * 100:.2f}%")

# 10. CONFUSION MATRIX & CLASSIFICATION REPORT
# --------------------------------------------
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# 10.1 True labels and predicted labels
y_true = test_generator.classes                        # shape=(num_test_samples,)
y_prob = model.predict(test_generator, verbose=1)       # shape=(num_test_samples, num_classes)
y_pred = np.argmax(y_prob, axis=1)

# 10.2 Confusion Matrix
cm = confusion_matrix(y_true, y_pred, labels=list(range(NUM_CLASSES)))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=CLASSES)

plt.figure(figsize=(8, 6))
disp.plot(cmap='Blues', xticks_rotation='45')
plt.title("Test Set Confusion Matrix")
plt.show()

# 10.3 Classification Report (Precision, Recall, F1)
report = classification_report(y_true, y_pred, target_names=CLASSES, digits=4)
print("Classification Report:\n")
print(report)

# 11. SHOW SOME MISCLASSIFIED EXAMPLES
# ------------------------------------
import matplotlib.image as mpimg

mis_idx = np.where(y_pred != y_true)[0]
print(f"\nFound {len(mis_idx)} misclassified examples out of {len(y_true)}")

n_show = min(9, len(mis_idx))
if n_show > 0:
    plt.figure(figsize=(12, 12))
    for i, idx in enumerate(mis_idx[:n_show]):
        img_path   = test_generator.filepaths[idx]
        true_label = CLASSES[y_true[idx]]
        pred_label = CLASSES[y_pred[idx]]

        ax = plt.subplot(3, 3, i + 1)
        img = mpimg.imread(img_path)
        plt.imshow(img)
        plt.axis('off')
        ax.set_title(f"True: {true_label}\nPred: {pred_label}", fontsize=10)
    plt.suptitle("Sample Misclassifications", fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("No misclassifications found (100% accuracy!).")
```
